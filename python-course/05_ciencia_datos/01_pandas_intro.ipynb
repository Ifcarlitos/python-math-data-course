# 1. Introducción a Pandas

Pandas proporciona estructuras de datos rápidas, flexibles y expresivas diseñadas para facilitar el trabajo con datos relacionales o etiquetados.

## 1.1 Series y DataFrames

```python
import pandas as pd
import numpy as np

# Serie: Array unidimensional etiquetado
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print("Serie:\n", s)

# DataFrame: Estructura tabular (hoja de cálculo)
data = {
    'Nombre': ['Ana', 'Juan', 'Luis', 'Maria'],
    'Edad': [23, 30, 25, 28],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla']
}
df = pd.DataFrame(data)
print("\nDataFrame:\n", df)
```

## 1.2 Inspección de Datos
Métodos básicos para explorar un DataFrame.

```python
print(df.head(2)) # Primeras filas
print("\nInfo:")
df.info()         # Tipos de datos y memoria
print("\nDescripción Estadística:")
print(df.describe()) # Resumen estadístico de columnas numéricas
```

## 1.3 Selección e Indexación
Acceder a filas y columnas específicas.

```python
# Selección de columna
edades = df['Edad']

# Selección por etiqueta (loc) y por posición (iloc)
print("Fila de Juan (loc):\n", df.loc[1])
print("Elemento (0, 0) (iloc):", df.iloc[0, 0])

# Filtrado condicional
mayores_25 = df[df['Edad'] > 25]
print("\nMayores de 25:\n", mayores_25)
```
