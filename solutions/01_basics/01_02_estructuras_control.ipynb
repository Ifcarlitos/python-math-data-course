# 2. Estructuras de Control de Flujo

En este notebook exploraremos cómo controlar el flujo de ejecución de nuestros programas mediante condicionales y bucles. Estas herramientas son fundamentales para implementar algoritmos numéricos.

## 2.1 Condicionales (`if`, `elif`, `else`)

```python
x = 10

if x > 0:
    print("Positivo")
elif x < 0:
    print("Negativo")
else:
    print("Cero")
```

## 2.2 Bucles

### Bucle `for`
Ideal para iterar sobre secuencias (listas, rangos, etc.).

```python
# Iterar sobre una lista
datos = [1, 5, 8, 12]
suma = 0

for valor in datos:
    suma += valor

print(f"Suma total: {suma}")

# Usando range()
print("Primeros 5 cuadrados:")
for i in range(5):  # 0, 1, 2, 3, 4
    print(f"{i}^2 = {i**2}")
```

### Bucle `while`
Se ejecuta mientras una condición sea verdadera.

```python
contador = 5
while contador > 0:
    print(f"Cuenta regresiva: {contador}")
    contador -= 1
print("¡Despegue!")
```

## 2.3 List Comprehensions
Una característica "pythónica" poderosa para crear listas de forma concisa. Muy útil en limpieza de datos.

```python
# Método tradicional
cuadrados = []
for x in range(10):
    cuadrados.append(x**2)

# List Comprehension
cuadrados_lc = [x**2 for x in range(10)]

print(f"Iguales? {cuadrados == cuadrados_lc}")

# Con condicional
pares = [x for x in range(10) if x % 2 == 0]
print(f"Pares: {pares}")
```
