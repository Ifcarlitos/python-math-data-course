# 1. Introducción a Python y Entorno Científico

Bienvenido al primer notebook del curso. Aquí aprenderemos los fundamentos del lenguaje Python, esenciales para cualquier aplicación matemática o de ciencia de datos.

## 1.1 ¿Por qué Python?
Python es el lenguaje estándar de facto en ciencia de datos debido a:
- **Sintaxis limpia:** Se lee casi como pseudocódigo.
- **Ecosistema robusto:** Librerías como NumPy, Pandas, y Scikit-Learn.
- **Comunidad activa:** Miles de recursos y soluciones disponibles.

## 1.2 Variables y Tipos de Datos
En Python, todo es un objeto. Las variables son nombres que apuntan a estos objetos.

```python
# Asignación de variables
x = 10         # Entero (int)
y = 3.14159    # Flotante (float)
z = "Python"   # Cadena de texto (str)
is_active = True # Booleano (bool)

print(f"x: {x}, tipo: {type(x)}")
print(f"y: {y}, tipo: {type(y)}")
print(f"z: {z}, tipo: {type(z)}")
```

### Tipos Numéricos
- `int`: Números enteros de precisión arbitraria.
- `float`: Números de punto flotante (doble precisión estándar IEEE 754).
- `complex`: Números complejos.

```python
# Operaciones matemáticas básicas
a = 5
b = 2

print(f"Suma: {a + b}")
print(f"Resta: {a - b}")
print(f"Multiplicación: {a * b}")
print(f"División: {a / b}")       # Retorna float
print(f"División entera: {a // b}") # Retorna int (piso)
print(f"Módulo: {a % b}")         # Resto
print(f"Potencia: {a ** b}")
```

## 1.3 Estructuras de Datos Básicas
Python tiene estructuras integradas poderosas: Listas, Tuplas, Diccionarios y Conjuntos.

```python
# Listas: Mutables y ordenadas
mi_lista = [1, 2, 3, "cuatro", 5.0]
mi_lista.append(6)
print(f"Lista: {mi_lista}")
print(f"Primer elemento: {mi_lista[0]}")

# Tuplas: Inmutables y ordenadas
mi_tupla = (1, 2, 3)
# mi_tupla[0] = 5  # Esto daría error

# Diccionarios: Pares clave-valor
mi_dict = {"nombre": "Curso Python", "version": 3.12}
print(f"Diccionario: {mi_dict['nombre']}")
```
