# 2. Visualización de Datos con Matplotlib y Seaborn

La visualización es crucial para entender los datos. `matplotlib` es la librería base, mientras que `seaborn` ofrece una interfaz de alto nivel para gráficos estadísticos.

## 2.1 Gráficos Básicos con Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Seno(x)', color='blue', linewidth=2)
plt.title("Función Seno")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()
```

## 2.2 Subplots
Podemos crear múltiples gráficos en una sola figura.

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

axs[0].plot(x, np.sin(x), 'r')
axs[0].set_title('Seno')

axs[1].plot(x, np.cos(x), 'g')
axs[1].set_title('Coseno')

plt.tight_layout()
plt.show()
```

## 2.3 Introducción a Seaborn
Seaborn hace que los gráficos estadísticos sean más atractivos y fáciles de crear.
Mas adeltante veremos mas detalles.

```python
import seaborn as sns

# Configurar estilo
sns.set_theme(style="darkgrid")

# Datos de ejemplo de Seaborn
tips = sns.load_dataset("tips")

plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Propinas vs Cuenta Total")
plt.show()
```
