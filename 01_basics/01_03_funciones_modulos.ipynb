# 3. Funciones y Programación Modular

Las funciones nos permiten encapsular lógica reutilizable, un principio clave (DRY - Don't Repeat Yourself) en ingeniería de software.

## 3.1 Definición de Funciones

````python
def saludar(nombre, mensaje="Hola"):
    """
    Imprime un saludo personalizado.
    Args:
        nombre (str): Nombre de la persona.
        mensaje (str, opcional): Mensaje de saludo. Por defecto "Hola".
    """
    print(f"{mensaje}, {nombre}!")

saludar("Ana")
saludar("Luis", "Buenos días")
````

## 3.2 Retorno de Valores
Las funciones pueden devolver resultados para ser usados posteriormente.

````python
def cuadrado_y_cubo(x):
    return x**2, x**3  # Retorna una tupla automáticamente

cuad, cub = cuadrado_y_cubo(3)
print(f"Cuadrado: {cuad}, Cubo: {cub}")
````

## 3.3 Funciones Lambda
Funciones anónimas pequeñas, útiles para pasar como argumentos a otras funciones (por ejemplo, en `map` o `filter`).

````python
doble = lambda x: x * 2
print(doble(5))

# Uso en ordenamiento
puntos = [(1, 2), (3, 1), (5, 0)]
# Ordenar por la segunda coordenada
puntos.sort(key=lambda p: p[1])
print(f"Ordenados por y: {puntos}")
````
